```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('datos/produccion_historica.csv')
df['fecha'] = pd.to_datetime(df['fecha'])

# Crear visualizacin simple
plt.figure(figsize=(10, 6))
for pozo in df['pozo'].unique():
    datos_pozo = df[df['pozo'] == pozo]
    plt.plot(datos_pozo['fecha'], datos_pozo['barriles_diarios'], label=pozo)

plt.title('Produccin Diaria por Pozo')
plt.xlabel('Fecha')
plt.ylabel('Barriles por Da')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grfico simple
import matplotlib.pyplot as plt

dias = [1, 2, 3, 4, 5]
produccion = [1200, 1250, 1180, 1300, 1280]

plt.figure(figsize=(8, 5))
plt.plot(dias, produccion, 'b-o', linewidth=2, markersize=8)
plt.title('Produccin Semanal')
plt.xlabel('Da')
plt.ylabel('Barriles')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Cargar y visualizar datos
df = pd.read_csv('datos/produccion_historica.csv')
df['fecha'] = pd.to_datetime(df['fecha'])

# Grfico por pozo
df.groupby('pozo')['barriles_diarios'].mean().plot(kind='bar')
plt.title('Produccin Promedio por Pozo')
plt.ylabel('Barriles por Da')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
