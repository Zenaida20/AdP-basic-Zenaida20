```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('datos/produccion_historica.csv')
df['fecha'] = pd.to_datetime(df['fecha'])

# Crear visualización simple
plt.figure(figsize=(10, 6))
for pozo in df['pozo'].unique():
    datos_pozo = df[df['pozo'] == pozo]
    plt.plot(datos_pozo['fecha'], datos_pozo['barriles_diarios'], label=pozo)

plt.title('Producción Diaria por Pozo')
plt.xlabel('Fecha')
plt.ylabel('Barriles por Día')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico simple
import matplotlib.pyplot as plt

dias = [1, 2, 3, 4, 5]
produccion = [1200, 1250, 1180, 1300, 1280]

plt.figure(figsize=(8, 5))
plt.plot(dias, produccion, 'b-o', linewidth=2, markersize=8)
plt.title('Producción Semanal')
plt.xlabel('Día')
plt.ylabel('Barriles')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Cargar y visualizar datos
df = pd.read_csv('datos/produccion_historica.csv')
df['fecha'] = pd.to_datetime(df['fecha'])

# Gráfico por pozo
df.groupby('pozo')['barriles_diarios'].mean().plot(kind='bar')
plt.title('Producción Promedio por Pozo')
plt.ylabel('Barriles por Día')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
